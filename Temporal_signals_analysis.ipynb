```python
x_shape_list = np.load("data/x_shape_list.npy")
y_shape_list = np.load("data/y_shape_list.npy")
file_list = np.load("Data/file_list.npy")

ROI_list = []

X_ROI = 15
Y_ROI = 15

for idx in tqdm(range(len(file_list))):
    # loadi into numpy array
    signal_array = np.loadtxt(file_list[idx], delimiter=',')
    signal_array = signal_array[:, :].T

    # Show matrix size
    print(file_list[idx])
    print("array shape:", signal_array.shape)
    print(x_shape_list[idx], x_shape_list[idx][0]*x_shape_list[idx][1])
    print(y_shape_list[idx], y_shape_list[idx][0]*y_shape_list[idx][1])

    datacube = signal_array.reshape(x_shape_list[idx][1], x_shape_list[idx][0], 2048) # If using filtered signal
    # Check size
    print("Final datacube shape (x,y,time):", datacube.shape)



    ROI = datacube[datacube.shape[0]-X_ROI:datacube.shape[0]+X_ROI,
                datacube.shape[1]-Y_ROI:datacube.shape[1]+Y_ROI,
                :]

    ROI_list.append(ROI)

ROI_arr = np.array(ROI_list)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm

n_samples = 4#ROI_arr.shape[0]
T = ROI_arr.shape[-1]
colors = cm.get_cmap('tab20', n_samples)

# ── 1. Compute per-sample mean signals and find their peaks ──────────────────
mean_signals = []
for i in range(n_samples):
    signals = ROI_arr[i].reshape(-1, T)   # (225, T)
    mean_signals.append(signals.mean(axis=0))

mean_signals = np.array(mean_signals)     # (n_samples, T)
peak_indices = np.argmax(mean_signals, axis=1)   # one peak index per sample

# ── 2. Align every mean signal so its peak lands on a common reference ───────
ref_peak = int(peak_indices.mean())       # central reference (or pick a fixed index)

aligned_signals = []
for i, sig in enumerate(mean_signals):
    shift = ref_peak - peak_indices[i]
    aligned_signals.append(np.roll(sig, shift))   # circular shift

aligned_signals = np.array(aligned_signals)       # (n_samples, T)

# ── 3. Plot aligned individual mean signals + global mean ────────────────────
t = np.arange(T)

plt.figure(figsize=(14, 6))

for i in range(n_samples):
    color = colors(i)
    label = file_list[i].split("\\")[-1]

    # Aligned per-sample mean signal
    sig = aligned_signals[i]

    # Re-compute std on the aligned pixel-level signals for the variance band
    shift = ref_peak - peak_indices[i]
    pixel_signals = np.roll(ROI_arr[i].reshape(-1, T), shift, axis=1)
    std_signal = pixel_signals.std(axis=0)

    plt.plot(t, sig, color=color, linewidth=2.5, label=label)
    # plt.fill_between(t,
    #                  sig - std_signal,
    #                  sig + std_signal,
    #                  color=color, alpha=0.25)

# Global mean across all aligned sample means
# global_mean = aligned_signals.mean(axis=0)
# plt.plot(t, global_mean, color='black', linewidth=2,
#          linestyle='--', label='Global mean (aligned)')

# plt.axvline(ref_peak, color='grey', linestyle=':', linewidth=1.2,
#             label=f'Ref peak (t={ref_peak})')

plt.title("Peak-Aligned Mean Temporal Signal ± Std Dev (per sample)")
plt.xlabel("Time points")
plt.ylabel("Signal amplitude")
plt.legend(fontsize=9)
plt.tight_layout()
plt.show()

# ── 4. Sanity-check: print shift applied to each sample ─────────────────────
# print(f"Reference peak index : {ref_peak}")
# for i in range(n_samples):
#     label = file_list[i].split("\\")[-1]
#     shift = ref_peak - peak_indices[i]
#     print(f"  {label:40s}  peak @ {peak_indices[i]:4d}  →  shift {shift:+d}")
```

```python
plt.figure(figsize=(14, 6))

for i in range(n_samples):
    signals = ROI_arr[i].reshape(-1, T)      # (225, T)
    color    = colors(i)
    label    = file_list[i].split("\\")[-1]

    # ── Align every pixel signal to the common reference peak ───────────────
    shift           = ref_peak - peak_indices[i]          # reuse from previous block
    signals_aligned = np.roll(signals, shift, axis=1)     # (225, T)

    # All pixel signals – low opacity density cloud
    plt.plot(t, signals_aligned.T, color=color, alpha=0.03)

    # Aligned mean overlay
    mean_aligned = signals_aligned.mean(axis=0)
    plt.plot(t, mean_aligned, color=color, linewidth=2, label=label)

# Reference peak marker
# plt.axvline(ref_peak, color='grey', linestyle=':', linewidth=1.2,
#             label=f'Ref peak (t={ref_peak})')

plt.title("Peak-Aligned Temporal Signals per Sample (Low Opacity Overlay)")
plt.xlabel("Time points")
plt.ylabel("Signal amplitude")
plt.legend(fontsize=9)
plt.tight_layout()
plt.show()
```

```python


import matplotlib.pyplot as plt

# PLOT CSCAN TO FIND AND VERIFY THE IMPACT SITE

# Scanning resolutions spatial and temporal
resolution_mm = 0.3  # spatial resolution in mm
dt_ps = 0.020464     # time interval in picoseconds

# Create axes
num_y, num_x, num_timepoints = datacube.shape
x_axis_mm = np.arange(num_x) * resolution_mm
y_axis_mm = np.arange(num_y) * resolution_mm
time_axis = np.arange(num_timepoints) * dt_ps

def show_cscan(requested_time_ps):
    idx = np.abs(time_axis - requested_time_ps).argmin()
    actual_time = time_axis[idx]
    cscan = datacube[:, :, idx]

    # Flip data vertically to make origin at bottom left (test if this is correct, otherwise do not use!!)
    #cscan_flipped = np.flipud(cscan)

    plt.figure(figsize=(6,6))
    plt.imshow(cscan, cmap='jet', origin='lower',
               extent=[x_axis_mm[0], x_axis_mm[-1], y_axis_mm[0], y_axis_mm[-1]])
    plt.title(f"C-Scan at time ≈ {actual_time:.3f} ps (Index {idx})")
    plt.xlabel("X position (mm)")
    plt.ylabel("Y position (mm)")
    plt.colorbar(label='Signal amplitude')
    plt.tight_layout()
    plt.show()


# select time slice in ps in funct argument (start time is 0)
time_slice = 5
show_cscan(time_slice)
```
